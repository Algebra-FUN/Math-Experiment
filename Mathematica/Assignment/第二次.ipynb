## 5.利用微分求函数的近似值
### 利用一阶Taylor展开

$$
f(x)=f(x_0)+(x-x_0) f^{'}(x_0)+o(|x-x_0|)
$$

其中$f(x)$是未知的，$x_0$是靠近$x$的一点且$f(x_0)$已知，$f^{'}(x)$可算

所以可以用$f(x_0)+(x-x_0) f^{'}(x_0)$作为$f(x)$的近似值

```
Nf[f_,x0_,x_]=f[x0]+(x-x0)D[f[t],t]/.t->x0
```

(1) 计算 $\arcsin 0.51$

```
Nf[ArcSin,0.51,0.5]
```

(2) 计算 ${126}^{\frac13}$

```
f[x_] = x^(1/3);
N[Nf[f,126,125]]
Clear[f];
```

### 利用级数展开
`Series[f,{x,x0,n}}]`

(1) 计算 $\arcsin 0.51$

```
Normal[Series[ArcSin[x],{x,.5,5}]]/.x->.51
```

(2) 计算 ${126}^{\frac13}$

```
N[Normal[Series[x^(1/3),{x,125,5}]]/.x->126]
```
