## 1.

从原始函数$y=\frac1{1+x^2},x\in[-5,5]$,随机取二十个点$\{x_i, y_i\}$

```
y[x_]:=1/(1+x^2)
```

```
data = Map[Function[x,{x,y[x]}],RandomReal[{-5,5},20]];
data//MatrixForm
```

### (1)
在[-5, 5]的区间，求二十个点的插值多项式 P10(x)和P20(x)，x最高阶为10和20

```
PolyFit[data_,n_]:=Fit[data,Table[x^i,{i,0,n}],x]
```

```
P10 = PolyFit[data,10]
P20 = PolyFit[data,20]
```

### (2)

```
DataPointPlot = ListPlot[data,PlotStyle->{Red}];
FunctionPlot = Plot[y[x],{x,-5,5},PlotStyle->{Dashed,Green}];
FitPlot = Plot[{P10,P20},{x,-5,5}];
Show[DataPointPlot,FitPlot,FunctionPlot]
```

### (3)	观察两个插值曲线在该区间上对原始函数的逼近情况
可以看出插值曲线在$x\in(-2,2)$上逼近情况良好，P20效果要好于P10,但是在两侧表现不佳,为过拟合现象

### 异同分析

| 命令         | 功能             | 最优解性质 | 初值依赖性     | 解的类型   | 返回值            |
| ------------ | ---------------- | ---------- | -------------- | ---------- | ----------------- |
| MaxValue     | 求最值           | 全局       | 不依赖         | 精确解     | 最值              |
| Maximizate   | 求最值及最值点   | 全局       | 不依赖         | 精确解     | {最值，x->最值点} |
| FindMaxValue | 搜寻极值         | 局部       | 依赖初值的选取 | 近似数值解 | 极值              |
| FindMaximum  | 搜寻极值及极值点 | 局部       | 依赖初值的选取 | 近似数值解 | {极值，x->极值点} |

#### 最值与极值

最值在紧集（一般是闭集）内一定存在，但是极值不一定存在

最值不一定是极值，极值也不一定是最值

#### 精确解与数值解

精确解存在不一定可以表示，但数值解存在一定可算
