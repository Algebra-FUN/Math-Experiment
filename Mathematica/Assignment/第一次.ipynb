# 第9题
## 题目
画曲线$4x^2+9y^2=36$绕$Ox$轴旋转而得的曲面
## 解
转化为参数方程组：
$$
\begin{cases}
x=3\cos u\\
y=2\sin u\cos v\\
z=2\sin u\sin v
\end{cases}
\quad s.t.
\begin{cases}
u\in[0,2\pi]\\
v\in[0,2\pi]\\
\end{cases}
$$

```
ParametricPlot3D[{3Cos[t],2Sin[t]Cos[u],2Sin[t]Sin[u]},{t,0,2Pi},{u,0,2Pi},AxesLabel->{HoldForm[x],HoldForm[y],HoldForm[z]}]
```

或者利用`RevolutionPlot3D`

```
?RevolutionPlot3D
```

```
RevolutionPlot3D[{3Cos[t],2Sin[t],0},{t,0,2Pi},RevolutionAxis->{1,0,0},AxesLabel->{HoldForm[x],HoldForm[y],HoldForm[z]}]
```

# 第11题
## 题目
画曲面$z=\sqrt{a^2-x^2-y^2}$与$z=\sqrt{x^2+y^2}-a$所围立体的图形
## 解答
经过分析,其区域应为
$$
\begin{cases}
z<\sqrt{a^2-x^2-y^2}\\
z>\sqrt{x^2+y^2}-a
\end{cases}
$$

```
a=1;
RegionPlot3D[x^2+y^2+z^2<a^2&&z>Sqrt[x^2+y^2]-a,{x,-a,a},{y,-a,a},{z,-a,a}]
```
